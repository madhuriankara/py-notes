```python
#Display the sum of Price for each Calorie_Level
```

```python
import pandas as pd
```

```python
FILEPATH = 'food.csv'
```

```python
df = pd.read_csv(FILEPATH)
```

```python
df
```

```python
import numpy as np
```

```python
df['Calorie_Level'] = np.where(df['Calories'] < 200, 'Low',
                      np.where(df['Calories'] <= 400, 'Medium', 'High'))
```

```python
df['Calorie_Level']
```

```python
df
```

```python
result = df.groupby('Calorie_Level')[['Calories']].sum()
```
