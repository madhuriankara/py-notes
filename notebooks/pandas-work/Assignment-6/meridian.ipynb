```python
#Calculate the median price of all food items.
```

```python
import pandas as pd
```

```python
food = pd.read_csv('food.csv')
```

```python
food_sales = pd.read_csv('food_sales.csv')
```

```python
df = pd.merge(food, food_sales, on = 'Food_Name')
```

```python
df
```

```python
result = df.groupby(df['Category'])[['Price']].median()
```
